```python
import matplotlib.pyplot as plt
s = ['Psychology', 'Nano Technology', 'Medicine', 'Oncology', 'Anthropology']
c, s1, s2 = 70, [67, 45, 73, 71, 57], [43, 67, 68, 53, 85]
plt.plot(s, s1,'o-', label='Student 1')
plt.plot(s, s2,'o-', label='Student 2')
plt.axhline(y=c, color='r', linestyle='--', label='Cutoff')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Comparison of Scores in Different Subjects')
plt.legend()
plt.show()
```

```python
#An india vs australia is on live.india has given 268 as  target for australia.the runs for every 5 overs is given as follows.
india scores for every 5 overs India ={25,51,84,131,160,189,220,250,267,297}
Australia have completed only upto 25 overs
runs of australia for every 5 overs are Aus={15,41,94,110,151}
Plot overs vs runs for both the countries with India as a blue line and australia as a yellow line.give requires labels and legend.the marker for india is flag,the marker for australia is plus.Take overs from arange function create runs as arrays(scores should be in arrays)
```

```python
import matplotlib.pyplot as plt
import numpy as np
overs_india = np.arange(5, 51, 5)
runs_india = [25, 51, 84, 131, 160, 189, 220, 250, 267, 297]
overs_aus = np.arange(5, 26, 5)
runs_aus = [15, 41, 94, 110, 151]
plt.plot(overs_india, runs_india, 'b-', label='India', marker='s', linewidth=3)
plt.plot(overs_aus, runs_aus, 'y-', label='Australia', marker='+', linewidth=3)
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('India vs Australia Live Match')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
overs_india = np.arange(5, 51, 5)
runs_india = [25, 51, 84, 131, 160, 189, 220, 250, 267, 297]
overs_aus = np.arange(5, 26, 5)
runs_aus = [15, 41, 94, 110, 151]
plt.plot(overs_india, runs_india, 'b', label='India', marker='>', linewidth=3)
plt.plot(overs_aus, runs_aus, 'y-', label='Australia', marker='+', linewidth=3)
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('India vs Australia Live Match')
plt.grid()
plt.legend()
plt.xticks(np.arange(0, 51, 5))
plt.show()
```
