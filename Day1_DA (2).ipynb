```python
#union
set1 = {11, 22, 33}
set2 = {33, 54, 75}
union_set = set1 | set2
print(union_set)
```

```python
#difference
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = set1.difference(set2)
print(set3)
set4 = set1 - set2
print(set4)
```
